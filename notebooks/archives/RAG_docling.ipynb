# OCRIsation et nécessité

```python
import pymupdf  # pip install pymupdf

pdf = "../data/files_test/ocr_test.pdf"
path_files = Path.cwd() / ".." / "data" / "10-DRH"


def pdf_necessite_ocr(pdf_path: str, seuil_min: int = 50) -> bool:
    
    try:
        doc = pymupdf.open(pdf_path)
        
        # On teste les 3 premières pages (ou moins si le PDF est court)
        nb_pages_a_tester = min(3, len(doc))
        
        for i in range(nb_pages_a_tester):
            page = doc[i]
            texte = page.get_text()
            
            # Si on trouve du texte exploitable, pas besoin d'OCR
            if len(texte.strip()) >= seuil_min:
                doc.close()
                return False
        
        # Aucune page avec assez de texte → besoin d'OCR
        doc.close()
        return True
        
    except Exception as e:
        print(f"Erreur lors de l'analyse : {e}")
        # En cas de doute, on suppose qu'il faut l'OCR
        return True


# Sort uniquement les pdf/docx et pas les autres fichiers, et on ignore les fichiers dans les dossiers "archives"
doc_paths = sorted(
    [p for p in path_files.rglob("*") 
     if p.suffix in {".pdf", ".docx"} and "archives" not in str(p).lower()],
    key=lambda p: str(p).lower()
)


docs = []

for doc in doc_paths:
    result = pdf_necessite_ocr(doc)
    docs.append(result)

print(f"Nombre de documents nécessitant OCR : {docs.count(True)}")
print(f"Nombre de documents ne nécessitant pas d'OCR : {docs.count(False)}")
```

```python
#embeddings
chunks = splitter.split_documents(raw_docs)
texts = [doc.page_content for doc in chunks]

embeddings = OllamaEmbeddings(
    model="nomic-embed-text",
    base_url="http://localhost:11434",
)

#indexation FAISS

db = None
batch_size = 16

for i in range(0, len(chunks), batch_size):
    batch = chunks[i:i+batch_size]
    if db is None:
        db = FAISS.from_documents(batch, embeddings)
    else:
        db.add_documents(batch)

db.save_local("./faiss_index")

# Chargement de l'index FAISS
db = FAISS.load_local(
    "./faiss_index",
    embeddings,
    allow_dangerous_deserialization=True
)
```

## EVAL

```python
# eval retrieval RAG

question = "comment se connecter au wifi résidents ?"

eval_set = [
  {
    "question": "comment obtenir le code pour se connecter au wifi résidents ?",
    "gold_snippet": "demander à l'accueil de l'ehpad",
  },
  {
    "question": "que dois je faire sur téléphone ou tablette ?",
    "gold_snippet": "désactiver l'option 'utiliser une adresse mac aléatoire'",
  },
]

def recall_at_k(db, eval_set, k=5):
    hits = 0
    for e in eval_set:
        docs = db.similarity_search(e["question"], k=k)
        combined = " ".join(d.page_content.lower() for d in docs)
        if e["gold_snippet"].lower() in combined:
            hits += 1
    return hits / len(eval_set)

print("Recall@5:", recall_at_k(db, eval_set, k=5))
print("Recall@10:", recall_at_k(db, eval_set, k=10))
```
